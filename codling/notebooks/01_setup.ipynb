## 4. Verify CUDA and GPU
Check CUDA availability, GPU information, and memory.

```python
import torch

print("=" * 50)
print("CUDA AND GPU VERIFICATION")
print("=" * 50)

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    print(f"CUDA version: {torch.version.cuda}")
    print(f"cuDNN version: {torch.backends.cudnn.version()}")
    print(f"Number of GPUs: {torch.cuda.device_count()}")
    
    for i in range(torch.cuda.device_count()):
        print(f"\nGPU {i}:")
        print(f"  Name: {torch.cuda.get_device_name(i)}")
        print(f"  Memory: {torch.cuda.get_device_properties(i).total_memory / 1024**3:.2f} GB")
    
    print(f"\nCurrent GPU memory allocated: {torch.cuda.memory_allocated() / 1024**3:.2f} GB")
    print(f"GPU memory reserved: {torch.cuda.memory_reserved() / 1024**3:.2f} GB")
else:
    print("⚠️  CUDA not available - training will be very slow on CPU")
```

## 5. Verify Installation
Test that all packages can be imported correctly.

```python
import importlib

packages = [
    ("torch", torch),
    ("transformers", None),
    ("datasets", None),
    ("accelerate", None),
    ("numpy", None),
    ("tqdm", None),
    ("matplotlib", None),
]

print("=" * 50)
print("PACKAGE VERIFICATION")
print("=" * 50)

for pkg_name, pkg_obj in packages:
    try:
        if pkg_obj is None:
            pkg_obj = importlib.import_module(pkg_name)
        version = getattr(pkg_obj, "__version__", "unknown")
        print(f"✅ {pkg_name}: {version}")
    except ImportError as e:
        print(f"❌ {pkg_name}: {e}")
```

## 7. Quick GPU Test
Run a quick sanity check to verify GPU is working.

```python
import torch

print("=" * 50)
print("GPU SANITY CHECK")
print("=" * 50)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

if torch.cuda.is_available():
    x = torch.randn(1000, 1000).cuda()
    y = torch.randn(1000, 1000).cuda()
    z = torch.matmul(x, y)
    print(f"✅ Matrix multiplication result shape: {z.shape}")
    print(f"✅ GPU memory allocated: {torch.cuda.memory_allocated() / 1024**3:.4f} GB")
else:
    print("Running on CPU (not recommended for training)")
    x = torch.randn(1000, 1000)
    y = torch.randn(1000, 1000)
    z = torch.matmul(x, y)
    print(f"✅ Matrix multiplication result shape: {z.shape}")
```
